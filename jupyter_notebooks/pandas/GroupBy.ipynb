```python
def split_location(x_df):
    def split_location_series(string):
        city,state = string.split(",")
        return pd.Series(
            (city,state),
            index = 'city State'.split()
        )
    res = x_df["Location"].apply(split_location_series)
    x_df[res.columns] = res
    return x_df
    
    
```

```python
def split_location_series(string):
        city,state = string.split(",")
        return pd.Series(
            (city,state),
            index = 'City State'.split()
        )
```
